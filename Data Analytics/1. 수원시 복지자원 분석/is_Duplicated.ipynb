# 행복e음, 복지자원조사 기관 중복 확인

## 1. 복지자원조사

```python
import csv
```

```python
f = open('C:/Users/User/Desktop/Dataset/duplicate_rs.csv','rt')
duplicated_rs = csv.reader(f)
```

### 리스트 중복값 제거

```python
unique_rs = []
for v in duplicated_rs :
    if v not in unique_rs:
        unique_rs.append(v)
print(unique_rs)
```

### 리스트 공백 제거

```python
import re
space = re.compile(r'\s+')
unique_rs = re.sub(space, '', str(unique_rs))
print(unique_rs)
```

### 유니크 데이터 List 형태 반환

```python
list_unique_rs = unique_rs.split(',')
print(list_unique_rs)
```

## 3. 기관 중복 확인

```python
def intersect(a, b):
    return list(set(a) & set(b))
```

```python
intersection = intersect(list_unique_sp,list_unique_rs)
print(intersection)
```
